```python
(X_train, y_train), (X_test,y_test) = datasets.cifar10.load_data()
X_train.shape
y_train[4001][0]
X_train.shape
# X_train
```

```python
from sklearn import preprocessing 
# normalization simply convert the value ranging from 0-1
X_train=X_train/255.0
X_test=X_test/255.0
```

```python
#  the term flatten is used to convert ths shape in to 1-D 
```

```python
cnn = models.Sequential([
    layers.Conv2D(filters=32, kernel_size=(3, 3), activation='relu', input_shape=(32, 32, 3)),
    layers.MaxPooling2D((2, 2)),
    
    layers.Conv2D(filters=64, kernel_size=(3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    
    layers.Flatten(),
    layers.Dense(64, activation='relu'),
    layers.Dense(10, activation='softmax')
])
```

```python
# here optimzer is nothing but the way it reduce the cost and also the back propogation and speedu opeartion.
cnn.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
cnn.fit(X_train, y_train, epochs=10)
```
